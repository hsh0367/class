```python
import pandas as pd
```

```python
train = pd.read_csv("../titanic/data/train.csv")
train.head(2)
```

## CountPlot

```python
import seaborn as sns
%matplotlib inline
sns.countplot(data=train, x='Embarked',hue="Survived")
# hue를 기준으로 먼저 나눈다
```

```python
sns.countplot(data=train, x='Pclass', hue="Survived")
```

## Barplot

```python
sns.barplot(data=train, x='Pclass', y="Fare", hue="Survived")
# 표준편차 표시
```

## Pointplot

```python
sns.pointplot(data=train, x='Pclass', y="Fare", hue="Survived")
```

```python
sns.distplot(train["Fare"], hist=False)
```

```python
sns.distplot(train["Fare"])
```

```python
low = train[train["Fare"]<100]
sns.distplot(low["Fare"], hist=False)
```

```python
perish = low[low["Survived"] == 0]
survived = low[low["Survived"] == 1]

sns.distplot(perish["Fare"], hist=False, label = "perish")
sns.distplot(survived["Fare"], hist=False, label = "syrvived")
```

```python
sns.lmplot(data=train, x="Age", y="Fare", hue="Survived", fit_reg=False)
```

```python
sns.lmplot(data=low, x="Age", y="Fare", hue="Survived", fit_reg=False)
```
